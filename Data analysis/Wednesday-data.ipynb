```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# dataset being loaded 
file_path = "C:/Users/Admin/Desktop/Data analysis/olympics-economics.csv"
data = pd.read_csv(file_path)


sns.set(style="whitegrid")

fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Histogram for total medals
sns.histplot(data['total'], bins=20, kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Distribution of Total Medals')
axes[0, 0].set_xlabel('Total Medals')

# Histogram for GDP 
sns.histplot(data['gdp'], bins=20, kde=True, log_scale=(True, False), ax=axes[0, 1])
axes[0, 1].set_title('Distribution of GDP (Log Scale)')
axes[0, 1].set_xlabel('GDP (in billion USD)')

# Line graph 
sns.lineplot(x='country', y='total', data=data, ax=axes[1, 0], marker='o')
axes[1, 0].set_title('Total Medals by Country')
axes[1, 0].tick_params(axis='x', rotation=90)

# Scatter plot of total medals vs. GDP
sns.scatterplot(x='gdp', y='total', data=data, ax=axes[1, 1])
axes[1, 1].set_title('Total Medals vs. GDP')
axes[1, 1].set_xlabel('GDP (in billion USD)')
axes[1, 1].set_ylabel('Total Medals')


plt.tight_layout()
plt.show()
```
